# MatplotLib Tutorials
#### Matplotlib is poltting library for the python programming language and its numerical mathematics extensions is NumPy. It provides an Object oriented API for embedding plots into applicationusing general-purpose GUI Toolkits like TKinter, wxPython, Qtor GTK

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x= np.arange(0,10)
y= np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
## plotting using matplotlib &plt scatter
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 20')
plt.show()
```

```python
y=x*x
```

```python
#plt plot
plt.plot(x,y,'r*--',linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel("X-AXis")
plt.ylabel("Y-Axis")
plt.title("2D Diagram")
```

```python
plt.plot(x,y,'r--')
```

```python
#Creating Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'b*')
plt.subplot(2,2,4)
plt.plot(x,y,'yo')
```

```python
x=np.arange(1,10)
y=3*x+5
plt.title("MatplotLIB Demo")
plt.xlabel("X-Axis caption")
plt.ylabel("Y-Axis caption")
plt.plot(x,y)
plt.show()
```

```python
##Sine Curve
x=np.arange(0,4*np.pi, 0.1)
y=np.sin(x)
plt.title("Sine Wave Form")

#plot the points using matplotlib
plt.plot(x,y, "r")
plt.show()
```

```python
# sine & cosine curve
x=np.arange(0, 5*np.pi,0.1)
y_sin =np.sin(x)
y_cos =np.cos(x)

#setup subplot grid , height is 2 and width is 1
plt.subplot(2,1,1)

#make first plot
plt.plot(x, y_sin, 'r*--')
plt.title('sine')

#second subplot as active & make it
plt.subplot(2,1,2)
plt.plot(x, y_cos, 'g--')
plt.title("cosine")

plt.show()
```

# Bar PLot

```python
x=[2,4,8,10]
y=[11,5,16,9]
x2=[3,4,9,11]
y2=[1,6,15,7]
plt.bar(x,y, align= 'center')
plt.bar(x2,y2, color ='g')
plt.title("BAR GRAPH")
plt.ylabel("Y- Axis")
plt.xlabel("X- Axis")
plt.show()
```

# Histrograms

```python
a=np.array([22,87,5,43,56,73,55,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

```python
a=np.array([22,87,5,43,56,73,55,11,20,51,5,79,31,27])
plt.hist(a, bins=20)
plt.title("Histogram")
plt.show()
```

# Box Plot using MatplotLIb

```python
data= [np.random.normal(0,std,100) for std in range(1,4)]
#rectangular box plot
plt.boxplot(data, vert =True, patch_artist=True);
```

# PIE Chart

```python
#data to plot
labels='Python', 'C++', 'Ruby', 'Java'
sizes =[215,130,245,210]
colors= ['gold','yellowgreen','lightcoral','lightskyblue']
explode =(0.1, 0,0,0) #explode 1st slice

#plot
plt.pie(sizes, explode =explode, labels=labels, colors=colors, autopct= '%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```
